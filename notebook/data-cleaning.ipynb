```python
#handling null values
df1 = pd.read_csv("shipment_orders/orders_data.csv", na_values=['Not Available','unknown'])
df1['Ship Mode'].unique()
```

```python
df1.head(5)
```

```python
#renaming columns
df1.rename(columns= {'Profit' :'Unit_profit', 'Selling_Price': 'Unit_Selling_Price','Order Id' :'Order_Id', 'Order Date': 'Order_Date','Ship Mode':'Ship_Mode','Postal Code' :'Postal_Code', 'Product Id': 'Product_Id','Total Profit':'Total_Profit','Sub Category':'Sub_Category'}, inplace=True)
df1.head(2)
```

```python
df1['Selling_Price'] = df1['List Price'] - df1['List Price']*(df1['Discount Percent']/100)
df1['Selling_Price']
```

```python
df1.head(2)
```

```python
df1['Profit'] = round(df1['Selling_Price'] - df1['cost price'], 2)
df1['Profit']
```

```python
df1.head(5)
```

```python
#changing datatype of Order date column from object to datetime
df1['Order_Date'] = pd.to_datetime(df1['Order_Date'], format='%d-%m-%Y')
```

```python
df1.dtypes
```

```python
#dropping few columns
df1.drop(columns= ['cost price','List Price','Discount Percent'], inplace=True)
```

```python
df1.head(2)
```

```python
#calculating total profit in each order
df1['Total_Profit'] = df1['Quantity']*df1['Profit']
```
